# Import The Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, display
```

# Load the Image

```python
image_path = "D:\dogs.jpg"
```

 # Load the image to grayscale

```python
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

# Invert the grayscale image

```python
inverted_gray = cv2.bitwise_not(gray_image)
```

# Apply Gaussian blur to the inverted image

```python
blurred_image = cv2.GaussianBlur(inverted_gray, (21, 21), 0)
```

# Invert the blurred image

```python
inverted_blur_image = cv2.bitwise_not(blurred_image)
```

# Divide the grayscale image by the inverted blurred image

```python
sketch_image = cv2.divide(gray_image, inverted_blur_image, scale=256.0)
```

# Display The original image

```python
display(Image(filename=image_path,width=300))
```

# Display The Grayscape image

```python
_, buffer = cv2.imencode('.jpg', gray_image)

# Convert the encoded buffer to bytes
image_bytes = buffer.tobytes()

# Display the grayscale image using IPython.display with a specified width
display(Image(data=image_bytes, format='jpg', width=300))
```

# Display The Pencil sketch

```python
# Encode the sketch image to JPEG format in memory
_, buffer = cv2.imencode('.jpg', sketch_image)

# Convert the encoded buffer to bytes and display the sketch
display(Image(data=buffer.tobytes(), width=300))
```
